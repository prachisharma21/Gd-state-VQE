# Project: To map the phase space for Mixed-field Ising model  

Mixed-field Ising hamiltonian is given by 

\begin{equation}
H= J \sum_{\langle ij \rangle} Z_i Z_j + h_x \sum_i X_i + h_z \sum_i Z_i,
\end{equation}
where $J = +1$ and $J = -1$ corresponds to anti-ferromagnetic and ferromagnetic phase. $h_x$ and $h_z$ corresponds the transverse and longitudinal applied field. You can find more details about the model and VQE ansatz built here in https://arxiv.org/pdf/2404.09132 

Considered lattices in the code are based on the native qubit connective of the devices used for simulation for example quito and Guadalupe backends. The images are shown below
